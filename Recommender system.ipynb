```python
import ast 

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
        return l
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df.head(2)
```

```python
movies_df.iloc[0]['keywords']
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head(2)
```

```python
def convert_cast(text):
    return [i['name'] for i in ast.literal_eval(text)[:3]]

movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
movies_df.head(2)
```

```python
movies_df.iloc[0]['crew']
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
#Sam Worthington
#SamWorthington

def remove_space(words):
    if not isinstance(words, list):
        return []
    return [i.replace(" ", "") for i in words]

```

```python
movies_df['cast'] = movies_df['cast'].apply(remove_space)
movies_df['crew'] = movies_df['crew'].apply(remove_space)
movies_df['genres'] = movies_df['genres'].apply(remove_space)
movies_df['keywords'] = movies_df['keywords'].apply(remove_space)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(
        list(enumerate(similary[index])),
        reverse=True,
        key=lambda x: x[1]
    )

    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```
